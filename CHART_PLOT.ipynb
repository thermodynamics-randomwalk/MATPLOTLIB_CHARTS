```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
```

```python
'''更换数据位置路径'''
path=os.getcwd()
df=pd.read_excel('2020年6月末企业内部原因分布（单位：亿元）.xlsx',parse_dates=True)

%matplotlib inline
import matplotlib.style as style
import matplotlib.animation as ani
plt.style.use('seaborn-whitegrid')
plt.rcParams['font.sans-serif']=['Microsoft YaHei']
plt.rcParams['axes.unicode_minus']=False
fig=plt.figure(figsize=(20,10))
a=df['企业内部原因'].values
b=df['信贷余额'].values
plt.barh(a,b,height=0.5)
plt.xlabel('报警笔数',fontweight='bold',fontsize=15)
plt.yticks(fontsize=20)
plt.ylabel('企业内部原因',fontweight='bold',fontsize=15)
plt.title('2020年6月末企业内部原因分布（单位：亿元）',pad=20,fontweight='bold',fontsize=20)
for t1,t2 in zip(a,b): 
    plt.text(t2+5, t1 ,'%.2f'%t2,va='center',fontsize=15)
plt.text(0.85, 0.02, '数据来源：分行报送',fontsize=10,fontweight='bold',transform=fig.transFigure)
plt.savefig('2020年6月末企业内部原因分布（单位：亿元）.png',bbox_inches='tight')
```

```python
df2=pd.read_excel('2017年至2020年6月末企业内部原因分布（单位：亿元）.xlsx','Sheet2',parse_dates=True)

fig=plt.figure(figsize=(20,10))
for i in df2.columns[1:4]:
    plt.plot(df2['年份'],df2[i],label=i,linewidth=3)
    for t1,t2 in zip(df2['年份'],df2[i]): 
        plt.text(t1, t2+3 ,'%.1f'%t2,va='center',ha='left',fontsize=15)
for i in df2.columns[4:]:
    plt.plot(df2['年份'],df2[i],label=i,linewidth=3)
    for t1,t2 in zip(df2['年份'],df2[i]): 
        plt.text(t1, t2-3 ,'%.1f'%t2,va='center',ha='left',fontsize=15)
plt.xlabel('年份',fontweight='bold',fontsize=15)

plt.yticks(fontsize=10)
plt.ylim(10,390)
plt.xlim(2017,2020)
plt.ylabel('信贷余额',fontweight='bold',fontsize=15)
plt.legend(fontsize=15, loc='upper right', bbox_to_anchor=(1.05, 1.03))
plt.grid(True)
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.title('2017年至2020年6月末企业内部原因分布（单位：亿元）',pad=30,fontweight='bold',fontsize=20)
plt.text(0.8, 0.08, '数据来源：分行报送',fontsize=15,fontweight='bold',transform=fig.transFigure)
plt.xticks(df2['年份'],labels=['2017年','2018年','2019年','2020年6月'],fontsize=15)
plt.savefig('2017年至2020年6月末企业内部原因分布（单位：亿元）.png',bbox_inches='tight')
```

```python
df7=pd.read_excel('2017年至2020年6月末我行内部原因占比分布.xlsx',parse_dates=True)

%matplotlib inline
import matplotlib.style as style
import matplotlib.animation as ani
plt.style.use('seaborn-whitegrid')
plt.rcParams['font.sans-serif']=['Microsoft YaHei']
plt.rcParams['axes.unicode_minus']=False
from matplotlib import cm
fig=plt.figure(figsize=(12,6))
ax1_colors=cm.rainbow(np.arange(len(df7['原因']))/len(df7['原因']))
plt.pie(df7['占比'],labels=df7['原因'],autopct='%.2f%%',shadow=False,
        startangle=0,textprops={'fontsize':10},colors=ax1_colors)
plt.title('2017年至2020年6月末我行内部原因占比分布情况',fontsize=15,fontweight='bold',pad=10)
plt.savefig('2017年至2020年6月末我行内部原因占比分布情况.png',bbox_inches='tight')
```

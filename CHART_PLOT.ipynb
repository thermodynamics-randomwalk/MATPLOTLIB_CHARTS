```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
```

```python
''''''
path=os.getcwd()
df=pd.read_excel('20206.xlsx',parse_dates=True)

%matplotlib inline
import matplotlib.style as style
import matplotlib.animation as ani
plt.style.use('seaborn-whitegrid')
plt.rcParams['font.sans-serif']=['Microsoft YaHei']
plt.rcParams['axes.unicode_minus']=False
fig=plt.figure(figsize=(20,10))
a=df[''].values
b=df[''].values
plt.barh(a,b,height=0.5)
plt.xlabel('',fontweight='bold',fontsize=15)
plt.yticks(fontsize=20)
plt.ylabel('',fontweight='bold',fontsize=15)
plt.title('20206',pad=20,fontweight='bold',fontsize=20)
for t1,t2 in zip(a,b): 
    plt.text(t2+5, t1 ,'%.2f'%t2,va='center',fontsize=15)
plt.text(0.85, 0.02, '',fontsize=10,fontweight='bold',transform=fig.transFigure)
plt.savefig('20206.png',bbox_inches='tight')
```

```python
df2=pd.read_excel('201720206.xlsx','Sheet2',parse_dates=True)

fig=plt.figure(figsize=(20,10))
for i in df2.columns[1:4]:
    plt.plot(df2[''],df2[i],label=i,linewidth=3)
    for t1,t2 in zip(df2[''],df2[i]): 
        plt.text(t1, t2+3 ,'%.1f'%t2,va='center',ha='left',fontsize=15)
for i in df2.columns[4:]:
    plt.plot(df2[''],df2[i],label=i,linewidth=3)
    for t1,t2 in zip(df2[''],df2[i]): 
        plt.text(t1, t2-3 ,'%.1f'%t2,va='center',ha='left',fontsize=15)
plt.xlabel('',fontweight='bold',fontsize=15)

plt.yticks(fontsize=10)
plt.ylim(10,390)
plt.xlim(2017,2020)
plt.ylabel('',fontweight='bold',fontsize=15)
plt.legend(fontsize=15, loc='upper right', bbox_to_anchor=(1.05, 1.03))
plt.grid(True)
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.title('201720206',pad=30,fontweight='bold',fontsize=20)
plt.text(0.8, 0.08, '',fontsize=15,fontweight='bold',transform=fig.transFigure)
plt.xticks(df2[''],labels=['2017','2018','2019','20206'],fontsize=15)
plt.savefig('201720206.png',bbox_inches='tight')
```
